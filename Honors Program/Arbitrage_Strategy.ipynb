```python
# import libraries
import time
import datetime
import pandas as pd
import numpy as np

# define a function to calculate daily returns from 2019-10-01 to 2022-10-01
# inputs: ticker symbol (type:str), floor (maximum buy-in price, type:nbr)
# outputs: daily_returns (a list of daily returns), size (# of daily returns), mean(mean value), std(standard deviation)
def five_day_returns(ticker,floor):
    period1 = int(time.mktime(datetime.datetime(2019,10,1,23,59).timetuple()))
    period2 = int(time.mktime(datetime.datetime(2022,10,1,23,59).timetuple()))
    interval = '1d'

    query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'

    df = pd.read_csv(query_string)
    HP = 5
    five_day_returns = []
    n = len(df['Close'])
    for i in range(n-HP):
        if df['Close'][i] <= floor:
            five_day_return = 0.99*df['Close'][i+HP]/1.01*df['Close'][i]-1
            five_day_returns.append(five_day_return)
    new_lst = []
    for i in five_day_returns:
        if str(i) != 'nan':
            new_lst.append(i)
    five_day_returns = new_lst
    size = len(five_day_returns)
    mean = np.mean(five_day_returns)
    std = np.std(five_day_returns)
    return five_day_returns, size, mean, std
```

```python
five_day_returns('VAI-USD',0.95)
```

```python
d = {'name':[],'size':[],'return':[],'std':[]}
ticker_list = ['USDT-USD',
'USDC-USD',
'BUSD-USD',
'DAI-USD',
'USDP-USD',
'TUSD-USD',
'GUSD-USD',
'USDD-USD',
'USDN-USD',
'FEI-USD',
'USTC-USD',
'FRAX-USD',
'LUSD-USD',
'EURS-USD',
'VAI-USD',
'SUSD-USD',
'OUSD-USD',
'CUSD-USD',
'HUSD-USD',
'RSV-USD']
d['name']=[i for i in ticker_list]
d['size']=[five_day_returns(i,0.9)[1] for i in ticker_list]
d['return']=[five_day_returns(i,0.9)[2] for i in ticker_list]
d['std']=[five_day_returns(i,0.9)[3] for i in ticker_list]
df = pd.DataFrame(data = d)
```
