```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('2021.xlsx')
```

```python
df
```

```python
dict1 = {'us':[], 'euro':[], 'else':[]}
```

```python
port_lst = list(set([i for i in df['Port']]))
print(port_lst)
```

```python
for i in port_lst:
    if i in ['US','NEW YORK','UNITED STATES OF AMERICA','美国','USA']:
        dict1['us'].append(i)
    elif i in ['DENMARK','Germany','FRANCE','德国','拉脱维亚','AUSTRIA','瑞典', 'SWEDEN','荷兰','FLORENCE','ISTANBUL','MILAN','Istanbul','MOSCOW','VERONA','PARIS','RUSSIAN FEDERATION, THE','İSTANBUL','ITALY','UNITED KINGDOM','HAMBURG','丹麦','俄罗斯','英国','意大利']:
        dict1['euro'].append(i)
    else:
        dict1['else'].append(i)
```

```python
label_list = []
for i in df['Port']:
    if i in dict1['us']:
        label_list.append(0)
    elif i in dict1['euro']:
        label_list.append(1)
    else:
        label_list.append(2)
```

```python
df.insert(3,'country_label',label_list)
```

```python
df.sort_values(by = ['country_label'])
```

```python
df1 = df.loc[df['Terms of Payment'] != 'CFR']
```

```python
temp = df.loc[df['Terms of Payment'] == 'CFR']
df2 = temp.loc[df['Freight charge'].notnull()]
```

```python
result = pd.concat([df1,df2])
```
