```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
# load data
data = pd.read_excel('input data.xlsx')
df = pd.DataFrame(data)
df
```

```python
# data clean
df = df.copy().dropna(axis = 0, subset = ['Port', 'Shipment Method', 'Total Cost'])
df[['Cubic Metre','Kilogram','Packages']] = df[['Cubic Metre','Kilogram','Packages']].fillna(0)
df
```

```python
df.columns = df.columns.str.replace(" ","_")
df.columns
```

```python
print(len(pd.unique(df['Port'])))
```

```python

fit = smf.ols(formula = 'Total_Cost ~ C(Date) + C(Port) + C(Shipment_Method) + Packages + Kilogram + Cubic_Metre', data=df).fit() 

fit.summary()
```
